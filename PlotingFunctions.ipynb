# Importing libraries



```python
import matplotlib.pyplot as plt
import numpy as np
```

# Define inputs of X

```python
x = np.arange(-10,10.1,0.1)
```

# Ploting y = wx + b

```python
w,b = 5,7

y = w * x + b

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = w*x + b')
```

# Ploting y = x^2

```python
y = x**2

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = x^2')
```

## Ploting y = 1/ (1 + e^(-x))

```python
y = 1 / (1 + np.exp(-x))

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = 1/ (1 + e^(-x))')
```

# Ploting y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
y = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = (e^x - e^(-x)) / (e^x + e^(-x))')
```

# Ploting y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
w,b = .9,0.5

y1 = w*x + b
y2 = 1 / (1 + np.exp(-y1))

plt.plot(x,y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = g(f(x))')
```

# Ploting y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
w,b = .9,0.5
y1 = w*x + b

y2 = (np.exp(y1) - np.exp(-y1)) / (np.exp(y1) + np.exp(-y1))

plt.plot(x,y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = g(f(x))')
```
